```python
df["count"] = df["Character_Fix"].map(lambda x: df["Character_Fix"].tolist().count(x) )
```

```python
df
```

```python
df.shape
```

```python
df_char_sort = df.sort_values(by=['count'],ascending=False)
```

```python
df_char_sort
```

```python
fig = px.scatter(df, x="Character_Fix", y="count",size="count",hover_name="Episode",color_continuous_scale="Bluered", size_max=60, title= "Total Lines spoken by Character")
fig.show()
```

```python
df_unique = df_char_sort.drop_duplicates(subset=['count'])
```

```python
df_unique
```

```python
top10character = df_unique.nlargest(10, 'count')
```

```python
top10character
```

```python
fig = px.bar(top10character, x="Character_Fix", y="count",color='Season',hover_name="Episode",  title= "Top lines by Character speaking by Season")
fig.show()
```
